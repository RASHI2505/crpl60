# Imputing Missing Values

```python
!pip install sklearn
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer= SimpleImputer(strategy='mean')
```

```python
imputer.fit(input_df[input1_df])
```

```python
input_df[input1_df]=imputer.transform(input_df[input1_df])
```

```python
test_df[input1_df]=imputer.transform(test_df[input1_df])
```

# Encoding Categorical Columns

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# 1. Creating the encoder
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
```

```python
encoder.fit(input_df[input2_df])
```

```python
# 3. Generate column names for each category
encoded_cols = list(encoder.get_feature_names(input2_df))
len(encoded_cols)
```

```python
encoded_cols
```

```python
input_df[encoded_cols]=encoder.transform(input_df[input2_df])
```

```python
test_df[encoded_cols]=encoder.transform(test_df[input2_df])
```

```python
input_df
```

# Training and Validation Sets

```python
from sklearn.model_selection import train_test_split
```

```python
train_df,val_df,train_target, val_target=train_test_split(input_df[input1_df + encoded_cols], target_df,test_size=0.25, random_state=42)
```

```python
val_df
```

```python
train_df
```

```python
train_target
```

# Training a Linear Regression Model

```python
from sklearn.linear_model import Ridge
```

```python
from sklearn.metrics import mean_squared_error
```

```python
model=Ridge()
```

```python
model.fit(train_df, train_target)
```

```python
predictions=model.predict(train_df)
predictions
```

```python
def rmse(target_df,predictions):
  return np.sqrt(np.mean(np.square(target_df-predictions)))
```

```python
train_rmse= rmse(train_target, predictions)
```

```python
print('RMSE LOSS for training set = : {}'.format(train_rmse))
```

```python
val_pred= model.predict(val_df)
val_pred
```

```python
val_rmse = rmse(val_target, val_pred)
```

```python
print('RMSE LOSS for validation set = : {}'.format(val_rmse))
```
